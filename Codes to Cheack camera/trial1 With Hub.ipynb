## Stream With two cameras

```python
# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'MP4V')
out = cv2.VideoWriter('output.mp4', fourcc, 30.0, (720,  480))
```

```python
while True:
    ret ,frame = cap.read()
    gray_image = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # out.write(gray_image)

    cv2.imshow("Frame",gray_image)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    
cap.release()
# out.release()
cv2.destroyAllWindows()
    
```

```python
while True:
    ret1 ,frame1 = cap1.read()
    gray_image = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
    # out.write(gray_image)

    cv2.imshow("Frame",gray_image)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    
cap.release()
# out.release()
cv2.destroyAllWindows()
    
```

# stream both of the cameras at the same time

```python
while True:
    # Capture frame-by-frame from each camera
    ret, frame = cap.read()
    ret1, frame1 = cap1.read()

    if not ret or not ret1:
        print("Failed to grab frame from one of the cameras")
        break

    # Display the resulting frame from each camera
    cv2.imshow('Camera 1', frame)
    cv2.imshow('Camera 2', frame1)

    # Press 'q' to close the windows
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cap1.release()
cv2.destroyAllWindows()
```
